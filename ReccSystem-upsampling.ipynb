```python
dfs = pd.read_csv('filtered_users.csv')
```

```python
df = dfs.sample(500);
```

```python
df['trip'] = np.random.randint(1, 3, df.shape[0])
```

```python
df['typeTrip'] = np.random.randint(1, 3, df.shape[0])
```

```python
df['durationTrip'] = np.random.randint(1, 3, df.shape[0])
```

```python
df
```

```python
# upsample the data to some 150000
```

```python
dfDummiesTrip = pd.get_dummies(df['trip'], prefix = 'trip')
```

```python
dfDummiesTrip
```

```python
dfDummiesTrip.rename(columns={'trip_1': 'soloTrip', 'trip_2': 'FamilyTrip'}, inplace=True)
```

```python
dfDummiesTrip
```

```python
dfDummiesTypeTrip = pd.get_dummies(df['typeTrip'], prefix = 'typeTrip')
```

```python
dfDummiesTypeTrip
```

```python
dfDummiesTypeTrip.rename(columns={'typeTrip_1': 'Holiday', 'typeTrip_2': 'Business'}, inplace=True)
```

```python
dfDummiesTypeTrip
```

```python
dfDummiesDurationTrip = pd.get_dummies(df['durationTrip'], prefix = 'durationTrip')
```

```python
dfDummiesDurationTrip
```

```python
dfDummiesDurationTrip.rename(columns={'durationTrip_1': 'short', 'durationTrip_2': 'long'}, inplace=True)
```

```python
dfDummiesDurationTrip
```

```python
newdf= df.join(dfDummiesTrip)
```
